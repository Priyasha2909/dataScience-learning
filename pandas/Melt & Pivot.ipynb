```python
import pandas as pd
 
# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
 
df = pd.DataFrame(data)
 
# Display the DataFrame
print(df)
```

```python
# id_vars = columns which we want to keep it as it is
# value_vars = columns we want to melt
# var_name = contains value_vars columns values
# value_name = contains values of columns in value_vars

melted = df.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name="Subject",value_name="Scores")
```

```python
# Pivot - to reshape into wider format

melted.pivot(index="Name", columns="Subject", values="Scores")
```
