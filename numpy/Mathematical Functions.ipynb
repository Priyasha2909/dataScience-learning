```python
import numpy as np
```

```python
arr = np.array([12,30,50])
```

```python
np.mean(arr)   # Average of an array
```

```python
np.max(arr)
```

```python
np.min(arr)
```

```python
np.std(arr)  # standard deviation
```

```python
np.var(arr)  # variance of array
```

```python
np.sum(arr)
```

```python
np.percentile(arr,50)  # compute 50th percentile
```

```python
np.prod(arr) # multiply
```

```python
np.unique(arr)   # find unique elements of array
```
